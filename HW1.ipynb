### Problem 1. (Build simple fully connected layer model)
- Build FCN Model
- cifar10: shape(32, 32, 3)


  
figure1.png
  

```
import torch
import torch.nn as nn

class SimpleFCN(nn.Module):
    def __init__(self, num_classes):
        super(SimpleFCN, self).__init__()
        #### please write your code below here.
        self.fcn1 = nn.Linear(32*32*3, 1024)
        self.fcn2 = nn.Linear(1024, 256)
        self.fcn3 = nn.Linear(256, 64)
        self.fcn4 = nn.Linear(64, 16)
        self.last_fc = nn.Linear(16, num_classes)
        self.activate = nn.LeakyReLU()
        #### please write your code above here.
    def forward(self, x):
        #### please write your code below here.
        x = torch.flatten(x, 1)
        x = self.fcn1(x)
        x = self.activate(x)
        x = self.fcn2(x)
        x = self.activate(x)
        x = self.fcn3(x)
        x = self.activate(x)
        x = self.fcn4(x)
        x = self.activate(x)
        x = self.last_fc(x)

        output = x
        #### please write your code above here.
        return output

```

### Problem 2. (dataset code part)
- We will use cifar10 dataset.
- If you want use other dataset, you can change the code.

```
from torchvision import datasets, transforms
import torchvision

def dataset(is_train):

    transform = transforms.Compose([
                                   transforms.ToTensor(),
                                   transforms.Normalize((0.5, 0.5, 0.5), (0.5,0.5, 0.5))])

    dataset = torchvision.datasets.CIFAR10(root='./data',train=is_train,download=True,transform=transform)


    return dataset
```

```
## Model train
```

```
def accuracy(output, target, topk=(1,)):
    """Computes the precision@k for the specified values of k"""
    maxk = max(topk)
    batch_size = target.size(0)

    _, pred = output.topk(maxk, 1, True, True)
    pred = pred.t()
    correct = pred.eq(target.view(1, -1).expand_as(pred))

    res = []
    for k in topk:
        correct_k = correct[:k].view(-1).float().sum(0)
        res.append(correct_k.mul_(100.0 / batch_size))
    return res
```

```
class AverageMeter(object):
    """Computes and stores the average and current value"""
    def __init__(self):
        self.reset()

    def reset(self):
        self.val = 0
        self.avg = 0
        self.sum = 0
        self.count = 0

    def update(self, val, n=1):
        self.val = val
        self.sum += val * n
        self.count += n
        self.avg = self.sum / self.count
```

```
GPU_NUM = 0
lr = 1e-4
num_classes = 10
#### please write your code below here.
batch_size = 128
epochs = 100  # Recommend 50 but you can use other number
#### please write your code above here.


device = torch.device(f'cuda:{GPU_NUM}' if torch.cuda.is_available() else 'cpu')


train_dataset = dataset(is_train=True)
val_dataset = dataset(is_train=False)


train_loader = torch.utils.data.DataLoader(train_dataset,
                                            batch_size=batch_size,
                                            shuffle=True)

val_loader = torch.utils.data.DataLoader(val_dataset,
                                            batch_size=batch_size,
                                            shuffle=False)


model = SimpleFCN(num_classes=num_classes).to(device)

optimizer = torch.optim.SGD(model.parameters(), lr, momentum=0.9, weight_decay=5e-4)

criterion = nn.CrossEntropyLoss().to(device)

```

```
train_loss_arr = []
train_acc_arr = []

val_loss_arr = []
val_acc_arr = []


for epoch in range(epochs):

    model.train()

    losses = AverageMeter()
    accs = AverageMeter()

    for i, (data, target) in enumerate(train_loader):

        data = data.to(device)
        target = target.to(device)

        #### please write your code below here.

        output = model(data)
        loss = criterion(output, target)

        output.float()
        loss.float()

        prec1 = accuracy(output.data, target)
        prec1 = prec1[0]

        losses.update(loss.item(), data.size(0))
        accs.update(prec1.item(), data.size(0))





        #### please write your code above here.

        if i % 100 == 0:
            print('Epoch: [{0}][{1}/{2}]\t'
                  'Loss {loss.val:.4f} ({loss.avg:.4f})'.format(epoch, i, len(train_loader), loss=losses))


        optimizer.zero_grad()

        loss.backward()

        optimizer.step()

    train_loss_arr.append(losses.avg)
    train_acc_arr.append(accs.avg)
    print("train result: Loss: {}, Acc: {}\n".format(losses.avg, accs.avg))


    model.eval()
    with torch.no_grad():
        val_loss_sum = 0
        val_acc_sum = 0

        losses = AverageMeter()
        accs = AverageMeter()

        for i, (data, target) in enumerate(val_loader):
            data = data.to(device)
            target = target.to(device)

            #### please write your code below here.
            output = model(data)
            loss = criterion(output, target)

            output.float()
            loss.float()

            prec1 = accuracy(output.data, target)

            prec1 = prec1[0]
            losses.update(loss.item(), data.size(0))
            accs.update(prec1.item(), data.size(0))







            #### please write your code above here.


            if i % 100 == 0:
                print('Test: [{0}/{1}]\t'
                      'Loss {loss.val:.4f} ({loss.avg:.4f})\t'
                      'Prec@1 {accs.val:.3f} ({accs.avg:.3f})'.format(
                          i, len(val_loader), loss=losses, accs=accs))

        val_loss_arr.append(losses.avg)
        val_acc_arr.append(accs.avg)
        print("Validation result: Loss: {}, Acc: {}\n".format(losses.avg, accs.avg))
```
